```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio
from scipy.io import wavfile
```

# Original wav

```python
path = r"./" # 
fname = r"Original_StarWars.wav" #  

fs, data = wavfile.read(path + fname)


plt.figure(figsize = (12, 3))
plt.plot(data, lw = 1)
plt.title("Orignal")
plt.xlabel("sample")
plt.ylabel("data")
plt.xlim(0, len(data))

Audio(data, rate=fs)
```

# Without Saturation(Overflow)

```python
saturation_data = np.add(data, 15000)  # add 15000

plt.figure(figsize = (12, 3))
plt.plot(saturation_data, lw = 1)
plt.title("saturation")
plt.xlabel("sample")
plt.ylabel("added_data")
plt.xlim(0, len(saturation_data))

Audio(saturation_data, rate=fs)
```

## Saturation(Clipping)

```python
threshold = 32767 - 15000

clipping_data = np.where( data < threshold, data + 15000, 32767)



plt.figure(figsize = (12, 3))
plt.plot(clipping_data, lw = 1)
plt.title("clipping")
plt.xlabel("sample")
plt.ylabel("added_data")
plt.xlim(0, len(clipping_data))

Audio(clipping_data, rate=fs)
```
